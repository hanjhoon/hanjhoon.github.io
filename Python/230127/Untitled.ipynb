```python
import tqdm
import os
host = "https://www.starbucks.co.kr"
for x in tqdm.tqdm(result):
    if os.path.isdir("./star") == False:
        os.mkdir("./star")
    request.urlretrieve(host + x['defaultimage'], "./star/{}.jpg".format(x['s_name']))
    x['period'] = (datetime.now() - datetime.strptime(x['open_dt'], "%Y%m%d")).days
```
