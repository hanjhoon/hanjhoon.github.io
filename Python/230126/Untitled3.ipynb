```python
krx_url = "http://data.krx.co.kr/comm/bldAttendant/getJsonData.cmd"
```

```python
payload = {"bld" : "dbms/MDC/STAT/standard/MDCSTAT01501",
"locale" : "ko_KR",
"mktId" : "ALL",
"trdDd" : "20230126",
"share" : "1",
"money" : "1",
"csvxls_isNo" : "false",}
```

### pickle 저장

```python
with open("./krx.pkl", "rb" ) as f:
    krx = pickle.load(f)
```

```python
len(krx.json()['OutBlock_1'])
```

```python
import os
import tqdm
import csv

for x in tqdm.tqdm(krx.json()['OutBlock_1']):
    #print(x['ISU_SRT_CD'], x['ISU_ABBRV'])
    if os.path.isdir("./stock") == False:
        os.mkdir("./stock")
    rt = get_stock(x['ISU_SRT_CD'],1,2)
    f = open(f"./stock/{x['ISU_ABBRV']}.csv", 'w', encoding='utf-8', newline="")
    wr = csv.writer(f)
    wr.writerows(rt)
    f.close()
```
