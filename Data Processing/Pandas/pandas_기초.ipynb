- Pandas Series	vs Pandas DataFrame



- Pandas Series
 - One-dimensional
 - Homogenous – Series elements must be of the same data type.
 - Size-immutable – Once created, the size of a Series object cannot be changed.	Size-mutable. 

- Pandas DataFrame
 - Two-dimensional
 - Heterogenous – DataFrame elements can have different data types. Size-immutable.
 - Elements can be dropped or added in an existing DataFrame.

```python
mask1=iris['sepal_length']<5.0
mask2=iris['sepal_width']>3.0
mask= mask1 | mask2

iris.loc[mask]
```

### Concat


- #### DataFrame 결합


```python
df1 = pd.DataFrame({'A': ['A1', 'A2', 'A3'],
                    'B': ['B1', 'B2', 'B3'],
                    'C': ['C1', 'C2', 'C3']},
                   index=['ONE', 'TWO', 'THREE'])
print('df1',df1,sep='\n')

print('-'*30)
df2 = pd.DataFrame({'C': ['C2', 'C3', 'C4'],
                    'D': ['D2', 'D3', 'D4']},
                   index=['TWO', 'THREE', 'FOUR'])
print('df2',df2,sep='\n')
print('-'*30)
s1 = pd.Series(['X1', 'X2', 'X3'], index=['ONE', 'TWO', 'THREE'], name='X')
print('s1',s1,sep='\n')
print('-'*30)
s2 = pd.Series(['Y2', 'Y3', 'Y4'], index=['TWO', 'THREE', 'FOUR'], name='Y')
print('s2',s2,sep='\n')
```

```python
df3=pd.DataFrame(s1)
print(df1+df2+df3)
```

```python
df_concat = pd.concat([df1, df2])
print(df_concat)
```

```python
df_concat_multi = pd.concat([df1, df2, df1])
print(df_concat_multi)
```

```python
df_v = pd.concat([df1, df2], axis=0)
print(df_v)
print('-'*30)
df_h = pd.concat([df1, df2], axis=1)
print(df_h)
```

#### Series 결합


```python
s1 = pd.Series(['X1', 'X2', 'X3'], index=['ONE', 'TWO', 'THREE'], name='X')

s2 = pd.Series(['Y2', 'Y3', 'Y4'], index=['TWO', 'THREE', 'FOUR'], name='Y')

s_v = pd.concat([s1, s2])
print(s_v)
print(type(s_v))
print('-'*30)
s_h = pd.concat([s1, s2], axis=1)
print(s_h)
print(type(s_h))
```

#### DataFrame과 Series 결합


```python
#세로
df_s_v = pd.concat([df1, s2])
print(df_s_v)
print('-'*30)
#가로
df_s_h = pd.concat([df1, s2], axis=1)
print(df_s_h)
```

```python
ddd=pd.concat([df1,df2,s1, s2], axis=0)
print(ddd)
ddd=pd.concat([df1,df2, s1], axis=0)
print(ddd)
```

### Merge
> merge() : 첫 번째 파라미터로 left, 두 번째 파라미터로 right를 지정, 합치고 싶은 2개의 DataFrame를 설정

```python
import pandas as pd
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3'], 'b': ['b_1', 'b_2', 'b_3']})
df_ab = df_ab.set_index(['a'])

df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4'], 'c': ['c_1', 'c_2', 'c_4']})
print(pd.merge(df_ab, df_ac, right_on='a', left_index=True))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3'], 'b': ['b_1', 'b_2', 'b_3']})
df_ab = df_ab.set_index(['a'])

df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4'], 'c': ['c_1', 'c_2', 'c_4']})
df_ac = df_ac.set_index(['a'])
print(pd.merge(df_ab, df_ac,left_index=True, right_index=True))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']}) 

df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']}) 
print(pd.merge(df_ab, df_ac))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']}) 

df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})
print(df_ab.merge(df_ac))
```

 - on, left_on, right_on

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']}) 
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']}) 
                     
print(pd.merge(df_ab, df_ac, on='a'))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']}) 
                     
df_ac = pd.DataFrame({'a_': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_ac, left_on='a', right_on='a_'))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_ac = pd.DataFrame({'a_': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_ac_, left_on='a', right_on='a_').drop(columns='a_'))
```

- how

```python
# 내부 결합 (inner_join): how='inner'
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})

print(pd.merge(df_ab, df_ac, on='a', how='inner'))
```

```python
# 왼쪽 결합 (left_join): how='left'
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_ac, on='a', how='left'))
```

```python
# 오른쪽 결합 (right_join): how='right'
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_ac, on='a', how='right'))
```

```python
# 외부 결합 (outer_join): how='outer'
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_ac, on='a', how='outer'))
```

- indicator


```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})

print(pd.merge(df_ab, df_ac, on='a', how='inner', indicator=True))
print('---------------------------')
print(pd.merge(df_ab, df_ac, on='a', how='outer', indicator=True))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'c': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_ac, on='a', how='outer', indicator='indicator'))
```

- suffixes

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_b = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'b': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_b, on='a'))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']})
                     
df_b = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'b': ['c_1', 'c_2', 'c_4']})
                     
print(pd.merge(df_ab, df_b, on='a', suffixes=['_left', '_right']))
```

- 여러개 키를 지정

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']
                     , 'x':['x_2', 'x_2', 'x_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'b': ['c_1', 'c_2', 'c_4']
                     , 'x':['x_1', 'x_2', 'x_2']})
                     
print(pd.merge(df_ab, df_ac, on='a'))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']
                     , 'x':['x_2', 'x_2', 'x_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'b': ['c_1', 'c_2', 'c_4']
                     , 'x':['x_1', 'x_2', 'x_2']})
                     
print(pd.merge(df_ab, df_ac))
print('---------------------------')
print(pd.merge(df_ab, df_ac, on=['a', 'x']))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']
                     , 'x':['x_2', 'x_2', 'x_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'b': ['c_1', 'c_2', 'c_4']
                     , 'x':['x_1', 'x_2', 'x_2']})
                     
print(pd.merge(df_ab, df_ac, on='a'))
```

```python
# DataFrame 컬럼명이 다른 열을 키로 사용하고 싶은 경우 left_on, right_on에 리스트 형태로 컬럼명을 설정
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']
                     , 'x':['x_2', 'x_2', 'x_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'b': ['c_1', 'c_2', 'c_4']
                     , 'x_':['x_1', 'x_2', 'x_2']})
                     
print(pd.merge(df_ab, df_ac, left_on=['a', 'x'], right_on=['a', 'x_']))
```

```python
df_ab = pd.DataFrame({'a': ['a_1', 'a_2', 'a_3']
                     , 'b': ['b_1', 'b_2', 'b_3']
                     , 'x':['x_2', 'x_2', 'x_3']})
                     
df_ac = pd.DataFrame({'a': ['a_1', 'a_2', 'a_4']
                     , 'b': ['c_1', 'c_2', 'c_4']
                     , 'x':['x_1', 'x_2', 'x_2']})
                     
print(pd.merge(df_ab, df_ac, on=['a', 'x'], how='inner')) 
print('---------------------------')
print(pd.merge(df_ab, df_ac, on=['a', 'x'], how='left')) 
print('---------------------------')
print(pd.merge(df_ab, df_ac, on=['a', 'x'], how='right')) 
print('---------------------------')
print(pd.merge(df_ab, df_ac, on=['a', 'x'], how='outer')) 
```
